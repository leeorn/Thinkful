## Selection Sort
This is the implementation of selection sort (in place) in python. <br>
The algorithm works as follows:

- finds the smallest element in the array
- swap it with the value at the "first" location  

```python
import time
import random
```

```python
# Set seed.
random.seed(a=100)

# Create default list.
short_list = list(random.sample(range(1000000), 10))
long_list = list(random.sample(range(1000000), 10000))
```

```python
short_list
```

```python
def SelectionSort(ar):
    for i in range(len(ar)):
        # assumes the smallest is there
        smallest = ar[i]
        location = i
        
        # looks for a smaller number
        for y in range(i, len(ar)):
            # if found a value lower than the temp "smallest" save the value and its location
            if ar[y] < smallest:
                smallest = ar[y]
                location = y

        # if didn't find a new smallest; doesn't need to do anything.
        if location == i:
            continue
            
        # when exiting the if, we have the smallest value and its location; now, swap
        temp = ar[i]
        ar[i] = ar[location]
        ar[location] = temp
        
    return ar
```

```python
start_time = time.time()

result_short = SelectionSort(short_list)

print("--- %s seconds ---" % (time.time() - start_time))

result_short
```

```python
start_time = time.time()

result_short = SelectionSort(long_list)

print("--- %s seconds ---" % (time.time() - start_time))

long_list
```
