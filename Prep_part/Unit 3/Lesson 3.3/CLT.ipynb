```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# new populations with n (sample size) = 10, p (probability) = 0.2 or 0.5, k (how many times repeated the experiment) 
pop1 = np.random.binomial(10, 0.2, 10000)
pop2 = np.random.binomial(10, 0.5, 10000) 
```

```python
sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)
```

```python
plt.hist(pop1, alpha = 0.5, label = "First population")
plt.hist(pop2, alpha = 0.5, label = "Second population")
plt.legend(loc = 'upper right') 
plt.show
```

Question 1: 
Increase the size of your samples from 100 to 1000, then calculate the means and standard deviations for your new samples and create histograms for each. Repeat this again, decreasing the size of your samples to 20. What values change, and what remain the same?

If I increase the sample size, I assume that both populations will get closer to normal distribution look.
In contrast, if I decrease the sample size, the graph will look less like a normal distribution since the sample size will represent the entire distribution less well

```python
sample1_larger = np.random.choice(pop1, 1000, replace=True)
sample2_larger = np.random.choice(pop2, 1000, replace=True)

plt.hist(sample1_larger, alpha= 0.5, label = "First population")
plt.hist(sample2_larger, alpha= 0.5, label = "Second population")
plt.legend(loc='upper right') 
plt.show
```

```python
sample1_smaller = np.random.choice(pop1, 20, replace=True)
sample2_smaller = np.random.choice(pop2, 20, replace=True)

plt.hist(sample1_smaller, alpha= 0.5, label = "First population")
plt.hist(sample2_smaller, alpha= 0.5, label = "Second population")
plt.legend(loc='upper right') 
plt.show
```

Question 2:
Change the probability value (p in the NumPy documentation) for pop1 to 0.3, then take new samples and compute the t-statistic and p-value. Then change the probability value p for group 1 to 0.4, and do it again. What changes, and why?

In this case, we change the probability of something to occur. We get more toward a "fair" coin situation; where there is an equal chance for the binomial to be 1 or 0 (H or T). In other words, the blue shape should look more like the orange one.

```python
# p = 0.3
pop3 = np.random.binomial(10, 0.3, 10000)
sample3 = np.random.choice(pop3, 100, replace=True)

plt.hist(sample3, alpha= 0.5, label = "First population")
plt.hist(sample2, alpha= 0.5, label = "Second population")
plt.legend(loc='upper right') 
plt.show
```

```python
from scipy.stats import ttest_ind
print(ttest_ind(sample2, sample3, equal_var=False))
```

```python
# p = 0.4
pop4 = np.random.binomial(10, 0.4, 10000)
sample4 = np.random.choice(pop4, 100, replace=True)

plt.hist(sample4, alpha= 0.5, label = "First population")
plt.hist(sample2, alpha= 0.5, label = "Second population")
plt.legend(loc='upper right') 
plt.show
```

```python
print(ttest_ind(sample2, sample4, equal_var=False))
```

Question 3:
Change the distribution of your populations from binomial to a distribution of your choice. Do the sample mean values still accurately represent the population values?

According to the Central Limit Theorem as long as we take enough data, the mean of a sample will still represent the general mean. So the answer is yes.

```python
pop5 = np.random.logistic(10, 0.2, 10000)
pop6 = np.random.logistic(10, 0.5, 10000)

sample5 = np.random.choice(pop5, 100, replace=True)
sample6 = np.random.choice(pop6, 100, replace=True)

plt.hist(sample5, alpha= 0.5, label = "First population")
plt.hist(sample6, alpha= 0.5, label = "Second population")
plt.legend(loc='upper right') 
plt.show
```
