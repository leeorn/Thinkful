```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Counties growth

### compare countries growth rate

```python
df = pd.read_csv("countries.csv")
```

```python
us = df[df.country == "United States"]
```

```python
china = df[df.country == "China"]
```

```python
israel = df[df.country == "Israel"]
```

```python
canada = df[df.country == "Canada"]
```

```python
plt.plot(us.year, us.population / 10**6)
plt.plot(canada.year, canada.population / 10**6)
plt.plot(israel.year, israel.population / 10**6)
plt.plot(china.year, china.population / 10**6)
plt.legend(["United States", "Canada", "Israel", "China"])
plt.xlabel("year")
plt.ylabel("population in millions")
plt.show()
```
